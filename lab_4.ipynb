Затенение Гуро

```python
import math
from names import RABBIT, DEER, STORM_TROOPER, FOX
from core import ObjModel, CustomImage

rabbit_obj = ObjModel(RABBIT)
deer_obj = ObjModel(DEER)
storm_trooper_obj = ObjModel(STORM_TROOPER)
fox_obj = ObjModel(FOX)

img = CustomImage(1000, 1000)
rabbit_obj.rotate((math.pi, 0, 0))
rabbit_obj.shift((0.005, 0.045, 15.0))
rabbit_obj.draw(img, (100000, 100000))
img.save('pictures/models/rabbit_guro_shading.png')

img = CustomImage(1500, 1500)
deer_obj.rotate((math.pi, math.pi / 4, 0))
deer_obj.shift((0, 700, 10000))
deer_obj.draw(img, (7000, 7000))
img.save('pictures/models/deer_guro_shading.png')

img = CustomImage(1500, 1500)
storm_trooper_obj.rotate((math.pi, math.pi, 0))
storm_trooper_obj.shift((0, 0, 100))
storm_trooper_obj.draw(img, (30000, 30000))
img.save('pictures/models/storm_trooper_guro_shading.png')

img = CustomImage(1000, 800)
fox_obj.rotate((math.pi, math.pi / 4, 0))
fox_obj.shift((0, 40, 1000))
fox_obj.draw(img, (6000, 6000))
img.save('pictures/models/fox_guro_shading.png')
```

Анимация карамельного шлема штурмовика

```python
import math
from names import STORM_TROOPER,CARAMEL_TEXTURE
from core import ObjModel, CustomImage, CustomAnimation

ticks = 360
duration = 15

anim = CustomAnimation()
for tick in range(ticks):
    img = CustomImage(1500, 1500)
    storm_trooper = ObjModel(STORM_TROOPER, CARAMEL_TEXTURE)
    storm_trooper.rotate((math.pi, 2 * math.pi * tick / ticks, 0))
    storm_trooper.shift((0, 0, 100))
    storm_trooper.draw(img, (30000, 30000))
    anim.add_image(img)
    print((tick + 1) / ticks * 100, '%')
anim.save('pictures/animation/storm_trooper.gif', duration)
```

Анимация бумажного кролика

```python
import math
from names import RABBIT, PAPER_TEXTURE
from core import ObjModel, CustomImage, CustomAnimation

ticks = 360
duration = 15

anim = CustomAnimation()
for tick in range(ticks):
    img = CustomImage(1000, 1000)
    rabbit = ObjModel(RABBIT, PAPER_TEXTURE)
    rabbit.rotate((math.pi, 2 * math.pi * tick / ticks, 0))
    rabbit.shift((0.005, 0.045, 15.0))
    rabbit.draw(img, (100000, 100000))
    anim.add_image(img)
    print((tick + 1) / ticks * 100, '%')
anim.save('pictures/animation/rabbit.gif', duration)
```

Анимация мраморной лисы

```python
import math
from names import FOX, MARBLE_TEXTURE
from core import ObjModel, CustomImage, CustomAnimation

ticks = 360
duration = 15

anim = CustomAnimation()
for tick in range(ticks):
    img = CustomImage(1200, 800)
    fox = ObjModel(FOX, MARBLE_TEXTURE)
    fox.rotate((math.pi, 2 * math.pi * tick / ticks, 0))
    fox.shift((0, 40, 1000))
    fox.draw(img, (6000, 6000))
    anim.add_image(img)
    print((tick + 1) / ticks * 100, '%')
anim.save('pictures/animation/fox.gif', duration)
```
